```python
#self parameter

class student:
    def __init__(self, handler_name, pending):
        self.pending = pending
        self.handler = handler_name
        print(f"student for {self.handler} created.")
    def display_info(self):
        print(f"owner: {self.handler}")
        print(f"pending: sh.{self.pending:.2f}")
a = student("joan",50000)
print("-" * 30)
b = student("geof",80_000)  
b.display_info()    
```

```python
# object

class citizen:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.c = "object"
rod = citizen("rod",5)
print(f"rod's age: {rod.age} years old")
print(f"rod's:{rod.c}")

jona = citizen("jona",7)
print(f"\nnew object:{jona.name}")
print(f"jona's age:{jona.age}years old")
```

```python
class animals:
    def __init__(self, name, size):
        self.name = name
        self.size = size
elephant = animals("elephant", "biggest")  
cat = animals("cat","smallest")
print(f"the bigest animal is a{elephant.name} *{elephant.size}*.")
print(f"the smalllest animal is a {cat.name} *{cat.size}*.")
        
```
