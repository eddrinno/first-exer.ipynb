```python
import pandas as pd
```

```python
#Load data using pandas
data = pd.read_csv("sales_data.csv")
```

```python
#Calculate Total_sales = Units_sold* Unit price
data["Total_sales"] = data["Units_Sold"] * data["Unit_Price"]
```

```python
#Group data by Region and product to compute total sales
my_total =(
    data.groupby(["Region", "Product"])["Total_sales"]
    .sum()
    .reset_index()
)
```

```python
#Identify the top 3 best-selling products(overall)
best = (
    data.groupby("Product")["Total_Sales"]
    .sum()
    .sort_values(ascending = False)
    .head(3)
)
    
```

```python
print(my_total)
print("\n best 3:")
print(best)
```

```python
#Export summary to new csv
my_total.to_csv("Sales_summary.csv", index = False)
print("\n Summary saved to sales_summary.csv")
```
